## Some image utilities

```python
from ml4a import image

# load an image from file, with an optional resize
img = image.load_image('../../assets/dog.jpg', (320, 320))

# display it
image.display(img)

# load an image from the web, resize manually
img = image.load_image('https://upload.wikimedia.org/wikipedia/commons/thumb/f/f4/The_Scream.jpg/471px-The_Scream.jpg')
img = image.resize(img, (320, 480))
image.display(img)
```

### Get some info about the image

```python
w, h = image.get_size(img)
aspect_ratio = image.get_aspect_ratio(img)

print('size of last image:', w, h)
print('aspect ratio of last image:', aspect_ratio)
```

### Modify the image and save it

```python
# center square crop of the image
img = image.crop_to_aspect_ratio(img, 0.3)
image.display(img)
image.save(img, 'media/myImageCropped.jpg')
```

## Loading frames from a movie

```python
player = image.MoviePlayer('./media/fingers.mp4')
print("movie fps %d, duration %d sec, %d frames" % (player.fps, player.duration, player.num_frames))

frame = player.get_frame(150)
image.display(frame)
```
