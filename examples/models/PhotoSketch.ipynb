## Run PhotoSketch

```python
from ml4a import image
from ml4a.models import photosketch

# load an input image
img = image.load_image('https://cdn.britannica.com/67/19367-050-885866B4/Valley-Taurus-Mountains-Turkey.jpg')

# run photosketch
img_sketch = photosketch.run(img)

# display the original and sketch images
image.display([img, img_sketch])
```
