## Running torch-dreams

The simplest technique involves loading an image, and creating a config which has all the parameters of our desired optimization.

```python
from ml4a import image
from ml4a.models import torch_dreams

#torch_dreams.print_layers()

config = {
    'objective': {
        'layer': 'Mixed_6b',
        'channel': 119
    },
    'size': (250, 400),
    'iters': 20,
    'lr': 2e-4,
    'weight_decay': 1e-1,
    'iters': 50
}

input_img = image.monalisa()
output_img = torch_dreams.run(config, input_img)
image.display([input_img, output_img])
```

Looping through several more examples to see the variety, this time from a different layer.

```python
import random

output_imgs = []

config['objective']['layer'] = 'Mixed_6e'
for i in range(3):
    config['objective']['channel'] = random.randint(1,20)
    output_img = torch_dreams.run(config, input_img)
    output_imgs.append(output_img)

image.display(output_imgs)
```
